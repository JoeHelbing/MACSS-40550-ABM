```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# read in data from csv to pandas dataframe
df = pd.read_csv("./data/seg_model_batch_run_data.csv")

# check the data
df.head()
```

```python
# create a plot of number of agents vs. Avg pct similar neighbors
sns.set_style("whitegrid")
sns.set_context("talk")
sns.set_palette("colorblind")
sns.set_color_codes("colorblind")

# create a plot of number of agents vs. Avg pct similar neighbors where each run
# is a dot and the mean is a line with intolerance as hue
fig, ax = plt.subplots(figsize=(12, 8))
sns.scatterplot(
    x="num_agents",
    y="Avg pct similar neighbors",
    hue="intolerance",
    data=df,
    ax=ax,
)
ax.set_title("Average pct similar neighbors vs. number of agents")
ax.set_xlabel("Number of agents")
ax.set_ylabel("Average pct similar neighbors")
```
